#### Convert dataframe to numpy

```python
numpy_array = data_frame.to_numpy()
```





#### 2(a) What is the shape of the numpy array?

```python
numpy_array.shape
```

#### 2(b)Split the numpy array into two numpy arrays:

```python
X,Y = numpy_array[:150,0:4],numpy_array[:150,4:5]
```





#### 2(c) Print the unique elements of the labels array.

```python
np.unique(Y)
```

#### 2(d) Create a dictionary to map unique labels into numerical labels.

```python
map_dictionary = {'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
```

#### 2(e) Convert the labels array into an integer-valued array using the dictionary map.

```python
Y = np.vectorize(map_dictionary.get)(Y)
zero_index = Y == 0
one_index  = Y == 1
two_index  = Y == 2
```

#### 3(a) Answer : We can see from the plots that attribute 2 and 3  is the best pair for segregation of the class labels

#### 3(b)

```python
X_reduced=TSNE(n_components=2).fit_transform(X)
```

```python
plt.scatter(X_reduced[:150,0:1][zero_index],X_reduced[:150,1:2][zero_index],label="label 0")
plt.scatter(X_reduced[:150,0:1][one_index],X_reduced[:150,1:2][one_index],label="label 1")
plt.scatter(X_reduced[:150,0:1][two_index],X_reduced[:150,1:2][two_index],label="label 2")
plt.title("Scatter Plots")

plt.legend()
plt.show()
```

#### 3(b) Answer : The feature reduction using t-SNE has further improved the class segregation.
#### 3(c)

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
plt.scatter(X_pca[:150,0:1][zero_index],X_pca[:150,1:2][zero_index],label="label 0")
plt.scatter(X_pca[:150,0:1][one_index],X_pca[:150,1:2][one_index],label="label 1")
plt.scatter(X_pca[:150,0:1][two_index],X_pca[:150,1:2][two_index],label="label 2")
plt.title("Scatter Plots")

plt.legend()
plt.show()
```

#### 3(c) Answer : The clusters formed using PCA are dispersed but the  can be segregated into three classes

```python
fig, axs = plt.subplots(1, 1,figsize =(5,3),tight_layout = True) 
axs.hist(X[:150,0:1], bins = 20, label = "Attribute 0") 
plt.legend()
plt.show() 
fig, axs = plt.subplots(1, 1,figsize =(5,3),tight_layout = True) 
axs.hist(X[:150,1:2], bins = 20, label = "Attribute 1") 
plt.legend()
plt.show() 
fig, axs = plt.subplots(1, 1,figsize =(5,3),tight_layout = True) 
axs.hist(X[:150,2:3], bins = 20, label = "Attribute 2") 
plt.legend()
plt.show() 

fig, axs = plt.subplots(1, 1,figsize =(5,3),tight_layout = True) 
axs.hist(X[:150,3:4], bins = 20, label = "Attribute 3") 
plt.legend()
plt.show() 
```

#### Attribute 0  has positively skewed

#### Attribute 2 and 3 have two peaks which is like multimodal distribution
